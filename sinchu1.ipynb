TRANFORMATION

```python
import pandas as pd

# Fill missing values properly (avoid chained assignment warning)
enrollment_data_df.loc[:, "guardian_contact"] = enrollment_data_df["guardian_contact"].fillna("not available")

# Display cleaned data
print(enrollment_data_df)
```

```python
import pandas as pd
#convert first name to uppercase(use the correct column name)
students_data_df['first_name']=students_data_df['first_name'].str.upper()
#display cleaned data
students_data_df
```

```python
import pandas as pd
#convert last name to uppercase(use the correct column name)
students_data_df['last_name']=students_data_df['last_name'].str.upper()
#display cleaned data
students_data_df
```

```python
import pandas as pd
#convert first name to uppercase(use the correct column name)
teachers_data_df['first_name']=teachers_data_df['first_name'].str.upper()
#display cleaned data
teachers_data_df
```

```python
import pandas as pd
#convert last name to uppercase(use the correct column name)
teachers_data_df['last_name']=teachers_data_df['last_name'].str.upper()
#display cleaned data
teachers_data_df
```

```python
import pandas as pd 
# calculate age for date of birth
students_data_df['date_of_birth']=pd.to_datetime(students_data_df['date_of_birth'])
students_data_df['age']= (pd.to_datetime('today')-students_data_df['date_of_birth']).dt.days//365
students_data_df
```

```python
import pandas as pd
#calculate performance percentage
student_performance_data_df['percentage'] = (student_performance_data_df['marks_obtained']/student_performance_data_df['total_marks']) * 100
student_performance_data_df
```

```python
import pandas as pd
# create full name
students_data_df['full_name'] = students_data_df['first_name'] + '' + students_data_df['last_name']
students_data_df
```

```python
import pandas as pd
# create full name
teachers_data_df['full_name'] = teachers_data_df['first_name'] + '' + teachers_data_df['last_name']
teachers_data_df
```

```python
import pandas as pd
# Categorical Encoding for Gender
students_data_df[' gender_encoded'] = students_data_df['gender'].map({'Male': 1, 'Female': 0})
students_data_df
```

```python
import pandas as pd
# Aggregate Student Performance
# Calculate average percentage for each student
student_avg_performance = student_performance_data_df.groupby('student_id')['percentage'].mean().reset_index() 
student_avg_performance.rename(columns={'percentage':'avg_percentage'}, inplace=True)
student_avg_performance
```

```python
import pandas as pd 
# Merge average performance with students df  
students_data_df = pd.merge(students_data_df,student_avg_performance, on='student_id', how='left')
students_data_df
```

```python
import pandas as pd
# Extract Year from Enrollment Date
enrollment_data_df['enrollment_date'] = pd.to_datetime(enrollment_data_df['enrollment_date'])
enrollment_data_df ['enrollment_year'] = enrollment_data_df['enrollment_date'].dt.year
enrollment_data_df
```

```python
import pandas as pd
# Add School Details to Students
# Merge students df with schools df to add school_name and district
students_data_df = pd.merge(students_data_df, schools_data_df[['school_id', 'school_name', 'district']], on='school_id', how='left')
students_data_df
```

```python
import pandas as pd
# Filter Active Enrollments
active_enrollment = enrollment_data_df[enrollment_data_df['status']=='Active']
active_enrollment
```

```python
import pandas as pd 
# Save Transformed Data to New CSV Files (Optional)
students_data_df.to_csv('transformed_students.csv',index=False)
student_performance_data_df.to_csv('transformed_student_performance.csv',index=False)
enrollment_data_df.to_csv('transformed_enrollment.csv',index=False)
students_data_df
student_performance_data_df
enrollment_data_df
```
